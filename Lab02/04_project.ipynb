```python
async def fetch_weather(client, city="London"):
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
    resp = await client.get(url)
    data = resp.json()
    return {
        "city": city,
        "temp": data["main"]["temp"],
        "desc": data["weather"][0]["description"]
    }
```

```python
async def main():
    async with httpx.AsyncClient() as client:
        news_task = asyncio.create_task(fetch_news(client))
        weather_task = asyncio.create_task(fetch_weather(client, "Tehran"))

        news, weather = await asyncio.gather(news_task, weather_task)

        print("ðŸŒ¤ Weather:")
        print(f"{weather['city']}: {weather['temp']}Â°C, {weather['desc']}\n")
        print("ðŸ“° Top News Headlines:")
        for i, headline in enumerate(news, 1):
            print(f"{i}. {headline}")

```
